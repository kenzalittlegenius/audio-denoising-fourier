$$

1. Dans le cas particulier $N=4$ démontrer le théorème 1.

Pour $N = 4$, on a $\zeta = e^{2i\pi/4} = e^{i\pi/2} = i$.

Les vecteurs de la base de Fourier discrète sont (en colonnes) :

$$
\varepsilon_0 = \begin{pmatrix}1\\1\\1\\1\end{pmatrix}, \quad
\varepsilon_1 = \begin{pmatrix}1\\i\\i^2\\i^3\end{pmatrix} = \begin{pmatrix}1\\i\\-1\\-i\end{pmatrix}, \quad
\varepsilon_2 = \begin{pmatrix}1\\i^2\\i^4\\i^6\end{pmatrix} = \begin{pmatrix}1\\-1\\1\\-1\end{pmatrix}, \quad
\varepsilon_3 = \begin{pmatrix}1\\i^3\\i^6\\i^9\end{pmatrix} = \begin{pmatrix}1\\-i\\-1\\i\end{pmatrix}
$$

La matrice $E$ est donc :

$$
E = \begin{pmatrix}
1 & 1 & 1 & 1 \\
1 & i & -1 & -i \\
1 & -1 & 1 & -1 \\
1 & -i & -1 & i
\end{pmatrix}
$$

Sa conjuguée transposée est :

$$
E^* = \begin{pmatrix}
1 & 1 & 1 & 1 \\
1 & -i & -1 & i \\
1 & -1 & 1 & -1 \\
1 & i & -1 & -i
\end{pmatrix}
$$

**Calcul de $E^* E$ :**

Le coefficient $(E^* E)_{k,\ell}$ (ligne $k$, colonne $\ell$) est :

$$
(E^* E)_{k,\ell} = \sum_{n=0}^{3} (E^*)_{k,n} \, E_{n,\ell} = \sum_{n=0}^{3} \overline{\zeta^{kn}} \, \zeta^{\ell n} = \sum_{n=0}^{3} \zeta^{(\ell - k)n}
$$

**Si $k = \ell$ :** chaque terme vaut $1$, donc la somme vaut $4$.

**Si $k \neq \ell$ :** posons $m = \ell - k \not\equiv 0 \pmod{4}$, alors :
$$
\sum_{n=0}^{3} (i^m)^n = \frac{(i^m)^4 - 1}{i^m - 1} = \frac{i^{4m} - 1}{i^m - 1} = \frac{1 - 1}{i^m - 1} = 0
$$
car $i^{4m} = (i^4)^m = 1^m = 1$.

Ainsi $E^* E = 4 I_4$, ce qui donne $E^{-1} = \frac{1}{4} E^*$.

Donc $E$ est inversible et la famille $\mathcal{E}$ est une base de $\mathbb{C}^4$. ✓

2. Démontrer le Lemme 1, puis le théorème par la première méthode.



Soit $k_0 \in \mathbb{Z}$ et $N \in \mathbb{N}^*$.

**Cas 1 : $k_0 \equiv 0 \pmod{N}$**

Alors $\zeta^{k_0} = e^{2i\pi k_0 / N} = e^{2i\pi m} = 1$ pour un certain entier $m$.
Donc $\sum_{n=0}^{N-1} \zeta^{k_0 n} = \sum_{n=0}^{N-1} 1^n = N$. ✓

**Cas 2 : $k_0 \not\equiv 0 \pmod{N}$**

Alors $\zeta^{k_0} \neq 1$ (car $e^{2i\pi k_0/N} = 1 \iff N \mid k_0$).
On applique la formule de la somme géométrique :
$$
\sum_{n=0}^{N-1} \zeta^{k_0 n} = \frac{(\zeta^{k_0})^N - 1}{\zeta^{k_0} - 1} = \frac{\zeta^{k_0 N} - 1}{\zeta^{k_0} - 1} = \frac{e^{2i\pi k_0} - 1}{\zeta^{k_0} - 1} = \frac{1 - 1}{\zeta^{k_0} - 1} = 0
$$
car $e^{2i\pi k_0} = 1$ pour tout entier $k_0$. ✓

#### Démonstration du Théorème 1 par la première méthode

On calcule le coefficient $(k, \ell)$ du produit $E^* E$ :

$$
(E^* E)_{k,\ell} = \sum_{n=0}^{N-1} (E^*)_{k,n} \, E_{n,\ell} = \sum_{n=0}^{N-1} \overline{\zeta^{kn}} \cdot \zeta^{\ell n} = \sum_{n=0}^{N-1} \zeta^{(\ell-k)n}
$$

**D'après le Lemme 1 :**

- Si $k = \ell$ : $\ell - k = 0 \equiv 0 \pmod{N}$, donc la somme vaut $N$.
- Si $k \neq \ell$ : $\ell - k \in \{-(N-1), \ldots, -1, 1, \ldots, N-1\}$, donc $\ell - k \not\equiv 0 \pmod{N}$, et la somme vaut $0$.

Ainsi :
$$
E^* E = N \cdot I_N
$$

Donc $E$ est inversible et $E^{-1} = \frac{1}{N} E^*$.

Les colonnes de $E$ (les $\varepsilon_\ell$) sont donc $N$ vecteurs linéairement indépendants de $\mathbb{C}^N$ : ils forment une **base de $\mathbb{C}^N$**, appelée **base de Fourier discrète**. ✓

 3. &#x1F625; Démontrer le théorème à l’aide de la deuxième méthode, en suivant les étapes
suivantes :

(a)  Montrer que l'application linéaire $\mathcal{L}$ est injective puis bijective.


On définit :
$$
L : \mathbb{C}_{N-1}[X] \to \mathbb{C}^N, \quad P \mapsto (P(1), P(\zeta), \ldots, P(\zeta^{N-1}))
$$

**Injectivité :** Soit $P \in \ker(L)$. Alors $P$ s'annule aux $N$ points distincts $1, \zeta, \zeta^2, \ldots, \zeta^{N-1}$ (distincts car $\zeta$ est une racine primitive $N$-ième de l'unité, donc les $\zeta^k$ pour $k = 0, \ldots, N-1$ sont deux à deux distincts). Or $\deg(P) \leq N-1$. Un polynôme de degré $\leq N-1$ ayant $N$ racines est nécessairement le polynôme nul. Donc $\ker(L) = \{0\}$ et $L$ est injective.

**Bijectivité :** $\mathbb{C}_{N-1}[X]$ et $\mathbb{C}^N$ ont tous deux dimension $N$. Une application linéaire injective entre deux espaces de même dimension finie est bijective. ✓



(b) Nous allons maintenant calculer $\mathcal{L}^{-1}$ puis sa matrice dans les bases qui nous intéressent.

On pose pour $k=0, 1,\dots, N-1$, $\widetilde L_k(X)=\prod_{j\neq k}(X-\zeta^j)$. Montrer que $\widetilde L_k(\zeta^k)\neq 0$.


On définit $\tilde{L}_k(X) = \prod_{j \neq k} (X - \zeta^j)$.

Alors :
$$
\tilde{L}_k(\zeta^k) = \prod_{j \neq k} (\zeta^k - \zeta^j)
$$

Ce produit est non nul car pour tout $j \neq k$, $\zeta^k \neq \zeta^j$ (les puissances de $\zeta$ sont distinctes). ✓



(c) On pose $L_k(X)=\frac
{1}{\widetilde L_k(\zeta^k)}\widetilde L_k(X)$. Montrer que $\mathcal{L}^{-1}(v_0,\dots, v_{N-1})=\sum_{k=0}^{N-1}v_k L_k(X)$. 


On pose $L_k(X) = \dfrac{\tilde{L}_k(X)}{\tilde{L}_k(\zeta^k)}$.

Par construction, $L_k(\zeta^j) = \delta_{k,j}$ (vaut $1$ si $j = k$, $0$ sinon).

Soit $v = (v_0, \ldots, v_{N-1}) \in \mathbb{C}^N$. Posons $P(X) = \sum_{k=0}^{N-1} v_k L_k(X)$.

On a $P(\zeta^j) = \sum_k v_k L_k(\zeta^j) = \sum_k v_k \delta_{k,j} = v_j$, donc $L(P) = v$.

Et $\deg(P) \leq N-1$ car chaque $L_k$ est de degré $N-1$.

Ainsi $L^{-1}(v_0, \ldots, v_{N-1}) = \sum_{k=0}^{N-1} v_k L_k(X)$. ✓



(d) Montrer que $\prod_{j=0}^{N-1}(X-\zeta^j)=X^N-1$. En déduire que $$\zeta^k\widetilde L_k(X)=\frac{\left(\frac{X}{\zeta^k}\right)^N-1}{\frac{X}{\zeta^k}-1}=\sum_{p=0}^{N-1}\left(\frac{X}{\zeta^k}\right)^p.$$


Les $\zeta^j$ pour $j = 0, \ldots, N-1$ sont exactement les $N$ racines $N$-ièmes de l'unité, c'est-à-dire les racines du polynôme $X^N - 1$. Donc :
$$
\prod_{j=0}^{N-1}(X - \zeta^j) = X^N - 1
$$

On en déduit :
$$
\tilde{L}_k(X) = \prod_{j \neq k}(X - \zeta^j) = \frac{X^N - 1}{X - \zeta^k}
$$

En effectuant la division euclidienne de $X^N - 1$ par $X - \zeta^k$, ou en notant que :
$$
\frac{X^N - 1}{X - \zeta^k} = \frac{\left(\frac{X}{\zeta^k}\right)^N - 1}{\frac{X}{\zeta^k} - 1} \cdot \frac{1}{\zeta^{k(N-1)}} \cdot \frac{\zeta^{kN} - 1}{1}
$$

Plus directement, on utilise la factorisation de la somme géométrique :
$$
\frac{X^N - 1}{X - \zeta^k} = \sum_{p=0}^{N-1} \zeta^{k(N-1-p)} X^p \cdot \text{... }
$$

Ou mieux, en posant $Y = X/\zeta^k$ :
$$
X^N - 1 = (\zeta^k Y)^N - 1 = \zeta^{kN} Y^N - 1 = Y^N - 1 \quad (\text{car } \zeta^{kN} = 1)
$$
donc :
$$
\frac{X^N - 1}{X - \zeta^k} = \frac{Y^N - 1}{Y - 1} \cdot \frac{1}{\zeta^k} = \frac{1}{\zeta^k} \sum_{p=0}^{N-1} Y^p = \frac{1}{\zeta^k} \sum_{p=0}^{N-1} \left(\frac{X}{\zeta^k}\right)^p = \sum_{p=0}^{N-1} \frac{X^p}{\zeta^{k(p+1)}} \cdot \zeta^k
$$

En simplifiant proprement :
$$
\zeta^k \tilde{L}_k(X) = \zeta^k \cdot \frac{X^N-1}{X - \zeta^k} = \frac{\left(\frac{X}{\zeta^k}\right)^N - 1}{\frac{X}{\zeta^k} - 1} = \sum_{p=0}^{N-1} \left(\frac{X}{\zeta^k}\right)^p
$$

ce qui est bien le résultat demandé. ✓



(e) En déduire que $\widetilde L_k(\zeta^k)=N\zeta^{-k}$ puis que
$$L_k(X)=\frac{1}{N}\sum_{p=0}^{N-1}\frac{1}{\zeta^{kp}}X^p.$$
Conclure sur les coefficients de $E^{-1}$ et montrer que $E^{-1}=\frac{1}{N}E^*$.


En évaluant en $X = \zeta^k$ :
$$
\zeta^k \tilde{L}_k(\zeta^k) = \sum_{p=0}^{N-1} \left(\frac{\zeta^k}{\zeta^k}\right)^p = \sum_{p=0}^{N-1} 1 = N
$$

Donc $\tilde{L}_k(\zeta^k) = N \zeta^{-k}$.

Ainsi :
$$
L_k(X) = \frac{\tilde{L}_k(X)}{\tilde{L}_k(\zeta^k)} = \frac{\tilde{L}_k(X)}{N\zeta^{-k}} = \frac{\zeta^k \tilde{L}_k(X)}{N} = \frac{1}{N} \sum_{p=0}^{N-1} \frac{X^p}{\zeta^{kp}}
$$

**Conclusion sur les coefficients de $E^{-1}$ :**

La matrice de $L^{-1}$ dans les bases canoniques est $E^{-1}$ (matrice de $L$ dans la base canonique de $\mathbb{C}_{N-1}[X]$ vers la base canonique de $\mathbb{C}^N$ est $E$, donc la matrice de $L^{-1}$ est $E^{-1}$).

Le coefficient $(p, k)$ de cette matrice correspond au coefficient de $X^p$ dans $L_k(X)$, qui est $\frac{1}{N} \frac{1}{\zeta^{kp}} = \frac{1}{N} \zeta^{-kp} = \frac{1}{N} \overline{\zeta^{kp}}$.

Or $(E^*)_{p,k} = \overline{E_{k,p}} = \overline{\zeta^{pk}} = \zeta^{-pk}$.

Donc le coefficient $(p,k)$ de $E^{-1}$ est $\frac{1}{N}(E^*)_{p,k}$, ce qui donne bien :
$$
\boxed{E^{-1} = \frac{1}{N} E^*}
$$
✓

Pour $\ell \in \{0, \ldots, N-1\}$, on a :
$$
\hat{x}_\ell = \sum_{n=0}^{N-1} x_n \, e^{-2i\pi \ell n/N}
$$

**Pour $\ell = 0$ :**
$$
\hat{x}_0 = \sum_{n=0}^{N-1} x_n \cdot 1 = \sum_{n=0}^{N-1} x_n \in \mathbb{R}
$$
car les $x_n$ sont réels. ✓

**Pour $\ell \in \{1, \ldots, N-1\}$ :**

Calculons $\overline{\hat{x}_\ell}$ :
$$
\overline{\hat{x}_\ell} = \overline{\sum_{n=0}^{N-1} x_n \, e^{-2i\pi \ell n/N}} = \sum_{n=0}^{N-1} \overline{x_n} \, e^{2i\pi \ell n/N}
$$

Puisque $x \in \mathbb{R}^N$, $\overline{x_n} = x_n$, donc :
$$
\overline{\hat{x}_\ell} = \sum_{n=0}^{N-1} x_n \, e^{2i\pi \ell n/N} = \sum_{n=0}^{N-1} x_n \, e^{-2i\pi (N-\ell) n/N}
$$
car $e^{2i\pi \ell n/N} = e^{-2i\pi(N-\ell)n/N}$ (puisque $e^{-2i\pi N n/N} = e^{-2i\pi n} = 1$).

Donc :
$$
\overline{\hat{x}_\ell} = \hat{x}_{N-\ell}
$$

ce qui est bien $\hat{x}_\ell = \overline{\hat{x}_{N-\ell}}$. ✓

(b) &#x1F625; Réciproquement, montrer que si $y\in \mathbb{C}^N$ vérifie $y_0\in\mathbb{R}$ et pour tout $l=1,2\dots, N-1$, $\overline{y_l}=y_{N-l}$ alors il existe $x\in\mathbb{R}^N$ tel que $y=\hat x$.

**Démonstration :**

Puisque la TFD est bijective, il existe un unique $x \in \mathbb{C}^N$ tel que $\hat{x} = y$, à savoir $x = \frac{1}{N} E y$ (TFD inverse). Il suffit donc de montrer que ce $x$ est en fait à valeurs réelles.

On a, pour $n \in \{0, \ldots, N-1\}$ :
$$
x_n = \frac{1}{N} \sum_{\ell=0}^{N-1} y_\ell \, e^{2i\pi \ell n/N}
$$

Calculons $\overline{x_n}$ :
$$
\overline{x_n} = \frac{1}{N} \sum_{\ell=0}^{N-1} \overline{y_\ell} \, e^{-2i\pi \ell n/N}
$$

On effectue le changement d'indice $\ell' = N - \ell$ (en convenant que $\ell' = 0$ quand $\ell = 0$) :

Pour $\ell = 0$ : le terme contribue $\overline{y_0} e^0 = y_0$ (car $y_0 \in \mathbb{R}$).

Pour $\ell = 1, \ldots, N-1$ : en posant $\ell' = N - \ell \in \{1, \ldots, N-1\}$,
$$
\overline{y_\ell} \, e^{-2i\pi \ell n/N} = y_{N-\ell} \, e^{-2i\pi \ell n/N} = y_{\ell'} \, e^{2i\pi \ell' n/N} \cdot e^{-2i\pi N n/N} = y_{\ell'} \, e^{2i\pi \ell' n/N}
$$
car $e^{-2i\pi N n/N} = e^{-2i\pi n} = 1$, et on a utilisé l'hypothèse $\overline{y_\ell} = y_{N-\ell}$.

Donc :
$$
\overline{x_n} = \frac{1}{N} \left( y_0 + \sum_{\ell'=1}^{N-1} y_{\ell'} e^{2i\pi \ell' n/N} \right) = \frac{1}{N} \sum_{\ell'=0}^{N-1} y_{\ell'} e^{2i\pi \ell' n/N} = x_n
$$

Ainsi $x_n \in \mathbb{R}$ pour tout $n$, donc $x \in \mathbb{R}^N$. ✓

# Question 6 

Un échantillonnage d'un morceau de 3mn de qualité standard donnera environ $N=8\cdot 10^6$ ($T=180s$ avec un échantillonnage à $f_s=44100 Hz$ ou $s^{-1}$). Un ordinateur standard de notre époque peut effectuer environ $200\cdot 10^9$ opérations par secondes (200 gigaFlops). \`A combien de temps peut on estimer le calcul des coefficients de Fourier discrets par la méthode naïve et par la transformée de Fourier rapide?



On considère un signal de durée $T = 180$ secondes échantillonné à la fréquence  
$f_e = 44100$ Hz.

### Calcul du nombre d’échantillons

Le nombre d’échantillons est donné par :
$$
N = T \times f_e
$$

Donc :
$$
N = 180 \times 44100 = 7\,938\,000 \approx 8 \times 10^6
$$


### Méthode naïve

La méthode naïve pour calculer la transformée de Fourier discrète nécessite un nombre d'opérations de l'ordre de :
$$
N^2
$$

Ainsi :
$$
N^2 = (8 \times 10^6)^2 = 6.4 \times 10^{13}
$$

On suppose que l’ordinateur peut effectuer :
$$
200 \times 10^9 \text{ opérations par seconde}
$$

Le temps de calcul est donc :
$$
\frac{6.4 \times 10^{13}}{200 \times 10^9} = 320 \text{ secondes}
$$



### Transformée de Fourier rapide (FFT)

La FFT nécessite un nombre d'opérations de l'ordre de :
$$
N \log_2(N)
$$

On utilise le changement de base :
$$
\log_2(N) = \frac{\log(N)}{\log(2)}
$$

On obtient approximativement :
$$
N \log_2(N) \approx 1.8 \times 10^8
$$

Le temps de calcul est alors :
$$
\frac{1.8 \times 10^8}{200 \times 10^9} = 9 \times 10^{-3} \text{ secondes}
$$



 Ainsi, La méthode naïve nécessite environ **320 secondes** (soit plus de 5 minutes),  
tandis que la FFT ne nécessite qu’environ **0.009 secondes**.

La FFT est donc **extrêmement plus rapide** que la méthode naïve.
